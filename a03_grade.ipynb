## Introduction
$LU$ decompositon is useful method to find solution of given matrix equation. $LU$ decompostion is bassed on Gauss theorem.

### Theory
As it written before to solve matrix equation we need to divide matrix $A$ to $L$ and $U$. We will solve these equations:

$A=L \times U$

$Ld=B$

$Ux = d$ 

Methods are described in Chapra and Canale book.  
As we said we use pivoting methods. In each iteration of Gauss algoithm pivoting method puts maximum element on column and tries decrease total error of equation, it is the main reason to use pivoting method. As well as substitution of rows in matrix $A$, we need to subsitute matrices $B$ and $L$. To calculate total error we will compare obtained matrix of $B$ with a true matrix $B$.
To perform iterative refinement of our solution by obtained matrix "X" we use method from Chapra and Canale book:
\begin{equation*}
\delta B = B - B_i
\end{equation*}
\begin{equation*}
A \times \delta X = \delta B
\end{equation*}
To calculate $A^{-1}$
We use method:
\begin{equation*}
A^{-1}_i = A \times I_i
\end{equation*}
Where $A^{-1}_i$ is $i^{th}$ column of $A^{-1}$ and $I_i$ is $i^{th}$ column of identity matrix $I$
From Chapra and Canale book we know how to find condition number:
\begin{equation*}
Cond[A] = ||A|| \times ||A^{-1}||
\end{equation*}
We use scaling for condition number to convert inverse matrix of $A$. 

To know error of solution we take a sum of absolute differences of true values of matrix and obtained matrix.

From this part you see performance of scaling on a system. We divide each element of row to maximum element of that row. So maximum element in each row will be less or equal $1$. If there is element very closer to zero we assume that pivoting method may be needed. 

```python
a1 = np.copy(a)

for i in range(n):
    mx = max(a1[i])
    a1[i] /= mx
w = False;
for i in range(n):
    for j in range(n):
        if (a1[i, j] < 1e-10):
            w = True
if (w):
    print("It may be needed to use pivoting")
```

You see output, so let's check later pivoting method and obtain results.  


There are main functions which are needed in our process:

Gauss - takes $A$ and $B$ matrices and calculate new matrix for $L$ and convert $A$ to $U$. Also we take parameter $piv$, if it's true it means wee need to apply pivoting method, otherwise we don't. 

Backsub and Forwsub are used to obtain final $X$. However Forsub doesn't return $X$, it calculates $d$, later you will see that we calculate $X$ separately.  

```python
def backsub(a,b):
    x = np.zeros(n,float)
    
    x[n-1] = b[n-1]/a[n-1,n-1]
    for i in range(n-2,-1,-1):
        sum = b[i]
        for j in range(i+1,n):
            sum -= a[i,j]*x[j]
        x[i] = sum / a[i,i]
    return x

            
def gauss(a, b, piv):
    for k in range(n-1):
        # Pivoting
        if (piv == True):
            p = k
            big = abs(a[k, k])
            # if a[k,k] < 1e-10:
                
            for  ii in range(k+1, n):
                dummy = abs(a[ii, k])
                if (dummy > big):
                    big = dummy
                    p = ii
            if (p != k):
                for jj in range(0, n):
                    dummy = a[p, jj]
                    a[p, jj] = a[k, jj]
                    a[k, jj] = dummy
                    dummy = L[p, jj]
                    L[p, jj] = L[k, jj]
                    L[k, jj] = dummy      
                dummy = b[p]
                b[p] = b[k]
                b[k] = dummy
                
        # Elimination
        for i in range(k+1,n):
            factor = a[i,k]/a[k,k]
            L[i, k] = factor
            for j in range(k,n):
                a[i,j] -= factor*a[k,j]
def forwsub(d, b):
    for i in range(n):
        sum = b[i]
        for j in range(0, i):
            sum -= L[i][j] * d[j]
        d[i] = sum
```

Here we solve $LU$ decomposition without and with pivoting. We obtained $x1$ and $x2$ are two solutions of non-pivoting and pivoting method with $b1$ and $b2$, respectively. 

```python
L = np.zeros((n,n),float);
U = np.zeros((n,n),float);

U = np.copy(a) 
x1 = np.zeros(n,float)

b1 = np.zeros(n,float);
b1 = np.copy(b);

#Solve LU decomposition without pivoting
gauss(U, b1, False)
for i in range(n):
    L[i, i] = 1;

d = np.zeros(n,float);
forwsub(d, b1)
x1 = backsub(U, d)

L = np.zeros((n,n),float)
U = np.copy(a) 
x2 = np.zeros(n,float)
#Solve LU decomposition with pivoting

b2 = np.zeros(n,float);
b2 = np.copy(b);

gauss(U, b2, True)
for i in range(n):
    L[i, i] = 1;
    
d = np.zeros(n,float);
forwsub(d, b2)
x2 = backsub(U, d)
```


We can see the error with pivoting method is $10$ times less than the error without pivoting solution. Let's mention that error in each case is small and close to zero. 

Now we can say that it is better to use pivoting. As we said in theory part, pivoting method decrease error of solution, possibly to reduce round-off error. Sometime we have factor closer to zero, and divison of another number by it will increase such number and we obtain bigger error in all calculations. On other hand pivoting requires more operation because of substitution in rows. Both of them works approximately for $O(n^3)$

From this part we calculate new error of new iteration. We mentioned it in theory part.

```python
#New iterations
L = np.zeros((n,n),float)
U = np.copy(a) 
delx = np.zeros(n,float)
delb = np.zeros(n,float)

for i in range(n):
    delb[i] = b[i] - b2[i]

gauss(U, delb, True)
for i in range(n):
    L[i, i] = 1;
    
d = np.zeros(n,float);
for i in range(n):
    sum = delb[i]
    for j in range(0, i):
        sum -= L[i][j]* d[j]
    d[i] = sum

delx = backsub(U, d)

xnew = np.zeros(n,float)
for i in range(n):
    xnew[i] = x2[i] + delx[i]
newerr = 0
bb = np.zeros(n,float)
for i in range(n):
    for j in range(n):
        bb[i] += a[i, j] * xnew[j]
    newerr += abs(bb[i] - b[i])
print("Total error not of new iteration ", newerr)
```

We expect that error will decrease and we got it. New error is less than previous. We assume that one iteration is enough for this matrix to show it. It doesn't play important role to use more iterations.
